### Annotate usage

```python
import matplotlib.pyplot as plt
y=[13,11,13,12,13,10,30,12,11,13,12,12,15]
plt.plot(y)
plt.ylim(ymax=32)
plt.annotate('Importatnt',xy=(10,30),arrowprops=dict(facecolor='blue',shrink=0.05));
plt.show()
```

### Sub plot

```python
fig=plt.figure()
ax1=fig.add_subplot(221)
ax1.plot([1,2,3],[4,5,6]);
ax2=fig.add_subplot(222)
ax2.plot([1,2,3],[6,5,4]);

ax3=fig.add_subplot(223)
ax3.plot([1,2,3],[4,5,6]);
ax4=fig.add_subplot(224)
ax4.plot([1,2,3],[6,5,4]);

plt.show()
```
